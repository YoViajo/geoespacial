Learn How to Visualize Geospatial Data in Jupyter using kepler.gl

Visualizing the Census Tracts of New York City using kepler.gl

```python
from keplergl import KeplerGl
import geopandas as gpd
import matplotlib.pyplot as plt
import matplotlib
```

Leer datos: 2010 census tract map combined with population data of New York City

```python
census=gpd.read_file('data.json')
print('Shape=>',census.shape)
census.head()
```

Mostrar nuestro conjunto de datos con plot de geopandas

```python
census.plot(figsize=(12,10))
```

## Visualizar zonas censales

```python
map2=KeplerGl(height=500,data={'Census Data':census})
map2
```

Guardar la visualización como HTML

```python
map2.save_to_html(file_name='nyc_census.html',read_only=True)
```
